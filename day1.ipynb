```python
def max_counts(text):
    counts = {}
    for i in text.split(''):
        if i in counts:
            counts[i] += 1
        else:
            counts[i] = 1
        return counts
```

```python
people = ['흥','흥','김','이','흥','김']
def max_count(people):
    counts = {}
    for i in people:
        if i in counts:
            counts[i] += 1
        else:
            counts[i] = 1
    return counts
counts = max_count(people)
print(counts)
print("가장큰키 :",max(counts))
print("가장큰값 :",max(counts.values()))
```

```python
training_data = [['연두', 3, '사과'],
                 ['노랑', 3, '사과'],
                 ['빨강', 2, '포도'],
                 ['빨강', 1, '포도'],
                 ['노랑', 3, '레몬']]
def fruit_counts(data):
    counts = {}
    for row in data:
        label = row[-1]
        if label not in counts:
            counts[label] = 1
        else:
            counts[label] += 1
    return counts
result = fruit_counts(training_data)
print(result)
```

```python
training_data = [['연두', 3, '사과'],
                 ['노랑', 3, '사과'],
                 ['빨강', 2, '포도'],
                 ['빨강', 1, '포도'],
                 ['노랑', 3, '레몬']]
def fruit_counts(data):
    counts = {}
    for row in data:
        color = row[0]
        if color not in counts:
            counts[color] = 1
        else:
            counts[color] += 1
    return counts
result = fruit_counts(training_data)
print(result)
```
